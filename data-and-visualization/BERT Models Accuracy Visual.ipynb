## Accuracy Visualization

```python
import pandas as pd
```

```python
#Note: distinct_words and words values were taken from count_vocab.py and confidence values were taken from mlm_test.py
BERT_data = pd.read_csv("BERT_data.csv")
BERT_data
```

```python
summed = BERT_data.groupby(['grammar', 'model'])['t/f'].sum().reset_index()
summed
```

```python
merged = pd.merge(BERT_data, summed, on=['grammar', 'model'])
merged
```

```python
merged = merged.rename(columns={'t/f_y': 'grammar_score'})
merged
```

```python
merged['accuracy'] =  merged['grammar_score']/3
merged
```

```python
grammar_performance = merged.groupby(['grammar', 'model'])['accuracy'].mean().reset_index()
grammar_performance
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
BERT = grammar_performance[grammar_performance['model'] == 'BERT']
CHILDES = grammar_performance[grammar_performance['model'] == 'CHILDES']
CHILD_TEXT = grammar_performance[grammar_performance['model'] == 'CHILD_TEXT']
WIKI_TEXT = grammar_performance[grammar_performance['model'] == 'WIKI_TEXT']

x_pos = np.arange(len(BERT))
labels = ['Articles', 'Opposites', 'Prepositions', 'Proverbs', 'Question Tags']

plt.figure(figsize=(10, 8))
plt.bar(x_pos - 0.4, BERT['accuracy'], width = 0.2, label='BERT')
plt.bar(x_pos - 0.2, CHILDES['accuracy'], width = 0.2, label='CHILDES')
plt.bar(x_pos + 0.2, CHILD_TEXT['accuracy'], width = 0.2, label='CHILD_TEXT')
plt.bar(x_pos + 0.4, WIKI_TEXT['accuracy'], width = 0.2, label='WIKI_TEXT')
plt.xticks(x_pos, labels)
plt.xlabel('Sentence Structures')
plt.ylabel('Accuracy')
plt.title('Accuracy of Each Model on Different Sentence Structures')
plt.legend()
plt.show()
```
